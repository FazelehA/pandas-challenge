```python
# Dependencies and Setup
import pandas as pd
```

```python
# Files to Load 
school_data_path = "Resources/schools_complete.csv"
student_data_path = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_path)
student_data = pd.read_csv(student_data_path)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

## District Summary

```python
#Calculate Totals, Averages and %s to create a DataFrame with summary of information

Total_schools=len(school_data_complete["school_name"].unique())
Total_students=len(school_data_complete["Student ID"])
Total_budget=school_data["budget"].sum()
Average_math_score=school_data_complete["math_score"].mean()
Average_reading_score=school_data_complete["reading_score"].mean()

Total_passing_math=len(school_data_complete.loc[school_data_complete["math_score"] >= 70])
Total_passing_reading=len(school_data_complete.loc[school_data_complete["reading_score"] >= 70])
Total_pass = len(school_data_complete.loc[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)])
                 
percentage_passing_math = (Total_passing_math/Total_students)*100
percentage_passing_reading = (Total_passing_reading/Total_students)*100
percentage_pass=(Total_pass/Total_students)*100

District_summary= pd.DataFrame({"Total Schools":Total_schools, "Total Students":[Total_students],
                                "Total Budget":Total_budget, "Average Math Score":Average_math_score, 
                                "Average Reading Score":Average_reading_score, "% Passing Math":percentage_passing_math, 
                               "% Passing Reading":percentage_passing_reading, "% Overall Passing":percentage_pass})
District_summary
```

```python
#Apply formatting
District_summary["Total Budget"] = District_summary["Total Budget"].map("${:,.0f}".format)
District_summary["Average Math Score"] = District_summary["Average Math Score"].map("{:,.2f}".format)
District_summary["Average Reading Score"] = District_summary["Average Reading Score"].map("{:,.2f}".format)
District_summary["% Passing Math"] = District_summary["% Passing Math"].map("{:,.2f}".format)
District_summary["% Passing Reading"] = District_summary["% Passing Reading"].map("{:,.2f}".format)
District_summary["% Overall Passing"] = District_summary["% Overall Passing"].map("{:,.2f}".format)
District_summary
```

## School Summary

```python
#Create a reduced dataframe with relevant information
school_summary_df=school_data_complete[["student_name","school_name","reading_score","math_score","type","size","budget"]]

#Set school name as the index
grouped_summary=school_data.set_index(["school_name"])["type"]

#Calculate the total number of students in each school
school_count=school_summary_df["school_name"].value_counts()

#Calculate the total school budget and per student budget
school_budget=school_summary_df.groupby(["school_name"])["budget"].mean()
student_budget=school_budget/school_count

#Calculate average math and reading scores for each school
average_math=school_summary_df.groupby(["school_name"])["math_score"].mean()
average_reading=school_summary_df.groupby(["school_name"])["reading_score"].mean()

#Calculate % passing math,reading and total
Total_passing_math=school_summary_df.loc[school_summary_df["math_score"] >= 70].groupby("school_name").count()["student_name"]
percentage_math=(Total_passing_math/school_count)*100

Total_passing_reading=school_summary_df.loc[school_summary_df["reading_score"] >= 70].groupby("school_name").count()["student_name"]
percentage_reading=(Total_passing_reading/school_count)*100

#Calculate % overall pass
Total_pass =school_summary_df.loc[(school_summary_df["math_score"] >= 70) & (school_summary_df["reading_score"] >= 70)].groupby("school_name").count()["student_name"]
percentage_total=(Total_pass/school_count)*100

#Create new dataframe
school_summary=pd.DataFrame({"Type":grouped_summary,
    "Total Students":school_count, "Total school budget":school_budget,"Per student budget": student_budget,
                             "Average math score":average_math,"Average reading score":average_reading,
                             "% passing math":percentage_math,
                             "% passing reading":percentage_reading, "% overall passing":percentage_total})
#school_summary

#Apply formatting
school_summary["Total school budget"] = school_summary["Total school budget"].map("${:,.0f}".format)
#school_summary["Per student budget"] = school_summary["Per student budget"].map("${:,.2f}".format)

school_summary
```

## Highest-Performing Schools (by % Overall Passing)

```python
highest_performers=school_summary.sort_values(by=["% overall passing"],ascending=False).head(5)
highest_performers
```

## Lowest-Performing Schools (by % Overall Passing)

```python
highest_performers=school_summary.sort_values(by=["% overall passing"]).head(5)
highest_performers
```

## Math Scores by Grade

```python
#Calculate the average math score by grade

ninth_math=school_data_complete.loc[school_data_complete["grade"] == "9th"].groupby("school_name").mean()["math_score"]
tenth_math=school_data_complete.loc[school_data_complete["grade"] == "10th"].groupby("school_name").mean()["math_score"]
eleventh_math=school_data_complete.loc[school_data_complete["grade"] == "11th"].groupby("school_name").mean()["math_score"]
twelfth_math=school_data_complete.loc[school_data_complete["grade"] == "12th"].groupby("school_name").mean()["math_score"]

#Create DataFrame
math_score_by_grade=pd.DataFrame({"9th":ninth_math,"10th":tenth_math,"11th":eleventh_math, "12th":twelfth_math})
math_score_by_grade
```

## Reading Scores by Grade

```python
#Calculate average reading scores by grade
ninth_reading=school_data_complete.loc[school_data_complete["grade"] == "9th"].groupby("school_name").mean()["reading_score"]
tenth_reading=school_data_complete.loc[school_data_complete["grade"] == "10th"].groupby("school_name").mean()["reading_score"]
eleventh_reading=school_data_complete.loc[school_data_complete["grade"] == "11th"].groupby("school_name").mean()["reading_score"]
twelfth_reading=school_data_complete.loc[school_data_complete["grade"] == "12th"].groupby("school_name").mean()["reading_score"]

#create dataframe
reading_score_by_grade=pd.DataFrame({"9th":ninth_reading,"10th":tenth_reading,"11th":eleventh_reading, "12th":twelfth_reading})
reading_score_by_grade
```

## Scores by School Spending

```python
#Reduced DataFrame
spending_df=school_summary[["Per student budget","Average math score","Average reading score","% passing math","% passing reading","% overall passing"]]

#reset index
spending_df=spending_df.reset_index(drop=True)

#Create bins
bins =[0,584.9,629.9,644.9,680]

#create labels for the bins
spending_range=["<$585","$585-630","$630-645","$645-680"]

#slice data and place in to bins and place data in new column
spending_df["Spending Ranges (Per Student)"]=pd.cut(spending_df["Per student budget"],bins, labels=spending_range, include_lowest=True)

#group by spending range
spending_df_2=spending_df.groupby("Spending Ranges (Per Student)")

#find the means
spending_df_2[["Average math score","Average reading score","% passing math","% passing reading","% overall passing"]].mean()
```

## Scores by School Size

```python
#Reduced DataFrame
size_df=school_summary[["Total Students","Average math score","Average reading score","% passing math","% passing reading","% overall passing"]]

#reset index
size_df=size_df.reset_index(drop=True)

#Create bins
bins =[0,999,1999,5000]

#create labels for the bins
school_size=["Small (<1000)","Medium (1000-2000)","Large (2000-5000)"]

#slice data and place in to bins and place data in new column
size_df["School Size"]=pd.cut(size_df["Total Students"],bins, labels=school_size, include_lowest=True)

#group by school size
size_df_2=size_df.groupby("School Size")

#find means
size_df_2[["Average math score","Average reading score","% passing math","% passing reading","% overall passing"]].mean()

```

## Scores by School Type

```python
#Reduced DataFrame
type_df=school_summary[["Type","Average math score","Average reading score","% passing math","% passing reading","% overall passing"]]

#reset index
type_df=type_df.reset_index(drop=True)

#groupby type
type_df_2=type_df.groupby("Type")

#find means
type_df_2[["Average math score","Average reading score","% passing math","% passing reading","% overall passing"]].mean()
```

## Observable Trends

### Looking at the scores by School Type data, Charter schools are seen to perform better in both maths and english with an overall higher passing rate of 90.4% comapred to 53.67% for District schools. 

### Small (less than 1000 students) and medium sized (between 1000 and 2000 students) schools performed similarly in math and english, however the performance for large sized schools (between 2000 and 5000 students) was significanlt worse with an overall pass rate of 58% compared to 90%.

### From the Scores by School Spending table it can be seen that schools with a lower spending range per student have a higher overall percentage pass rate on average.
